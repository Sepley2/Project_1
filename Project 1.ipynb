#Pre-Processing

```python
#Firstly, let's import Pandas
import pandas as pd
```

```python
#Import and read our CSV
poke_df = pd.read_csv("PokemonStats.csv")
```

```python
#Let's check and see if it worked
poke_df
```

```python
#Looking at our columns
poke_df.columns
```

```python
#Checking our data types
poke_df.dtypes
```

```python
#Alright, let's drop the columns we know we won't need and address our nulls
#For the purpose of my project, I will be dropping Height, Weight, and ID.
poke_df.drop('Height', axis=1, inplace=True)
#DroppingID
poke_df.drop('ID', axis=1, inplace=True)
```

```python
#Dropping Weight
poke_df.drop('Weight', axis=1, inplace=True)
```

```python
poke_df
```

```python
#Next, lets look at our dreaded nulls
poke_df.isnull().sum()
```

```python
#Our only nulls are Type2 of course. :( Let's just change that to No Secondary Type
poke_df["Type2"].fillna("No Secondary Typing", inplace = True)
```

```python
poke_df["Type2"]
```

```python
#Oops I made a mistake. No Secondary Typing looks kinda janky. Let's fix that.
poke_df['Type2'].replace({'No Secondary Typing': 'No Type'}, inplace=True)
```

```python
poke_df['Type2']
```

```python
#It worked! Now, we should have no null values. Let's recheck.
poke_df.isnull().sum()
```

```python
#Perfect! No more Nulls. I hate those guys
poke_df
```

```python
#I am being a perfectionist, so let's rename our Type1 and Type2 column to more accurately reflect how typing works
poke_df.rename(columns={'Type1': 'Primary Type', 'Type2': 'Secondary Type'}, inplace=True)
poke_df
```
